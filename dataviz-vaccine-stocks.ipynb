```python
#import libraries
from jupyterthemes import jtplot
jtplot.style(theme = 'monokai', context = 'notebook', ticks = True, grid = False) 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the stock prices data using pandas
df = pd.read_csv('/kaggle/input/covid-vaccine-companies-stock-data-from-2019/vaccine stocks.csv')

df
```

```python
#data cleaning - remove unnecessary columsn for analysis
columns_to_keep = ['Date','High_BioNTech', 'High_Moderna', 'High_Novavax', 'High_Astrazeneca',
                  'High_Inovio Pharmaceuticals','High_Sinovac', 'High_Johnson & Johnson']
stock_df = df[columns_to_keep]
```

```python
stock_df.info()
```

```python
#BioNTech stock prices
stock_df.plot(x = 'Date', y = 'High_BioNTech', label = 'BioNTech Stock Price', figsize =(15,10), linewidth=2)
plt.ylabel('Price[$]')
plt.title('Data Viz of C19 Vaxx - 2019 - 2021 Plot')
#plt.legend(loc ='upper center')
plt.grid()
```

```python
# Plotting each company's stock price
fig, axes = plt.subplots(nrows=7, ncols=1, figsize=(15, 20))

companies = ['High_BioNTech', 'High_Moderna', 'High_Novavax', 'High_Astrazeneca', 'High_Inovio Pharmaceuticals', 'High_Sinovac', 'High_Johnson & Johnson']

for i, company in enumerate(companies):
    ax = axes[i]
    company_name = company.replace('High_', '')  # Remove the 'High_' prefix
    plot_label = f'{company_name} Stock Price'
    stock_df.plot(y=company, label=plot_label, ax=ax, linewidth=2)
    ax.set_ylabel('Price [$]')
    ax.set_title(f'Stock Price of {company_name} - 2019 - 2021')
    ax.grid()

plt.tight_layout()
plt.show()
```

# PLOT SCATTERPLOT 

```python
#read data from two stocks
x = stock_df['High_BioNTech']
y = stock_df['High_Moderna']
```

```python
#create a scatterplot
plt.figure(figsize = (15,10))
plt.scatter(x,y)
plt.title('Stock Prices of BioNTech and Moderna - 2019 - 2021')
plt.grid()
```

# PLOT PIE CHART

```python
h_value_Bio = stock_df['High_BioNTech'].max()
h_value_Az = stock_df['High_Astrazeneca'].max()
h_value_Iv = stock_df['High_Inovio Pharmaceuticals'].max()
h_value_Sv = stock_df['High_Sinovac'].max()
h_value_JJ = stock_df['High_Johnson & Johnson'].max()
h_value_Mod = stock_df['High_Moderna'].max()
h_value_Nv = stock_df['High_Novavax'].max()
```

```python
# Use matplotlib to plot a pie chart 
values = [h_value_Sv, h_value_Bio, h_value_Az, h_value_JJ, h_value_Mod, h_value_Nv,h_value_Iv]
colors = ['g','r','y','b','m','yellow','magenta']
labels = ['Sinovac','BioNTech','Astrazeneca','J&J', 'Moderna', 'Novavax','Inovio']
explode = [0.1, 0, 0, 0, 0.1, 0, 0]


plt.figure(figsize = (10,10))
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('Vaccine Stocks [High] Prices')
```

# PLOT HISTOGRAMS

```python
mu = stock_df['High_Moderna'].mean()
sigma = stock_df['High_Moderna'].std()

num_bins = 40
plt.figure(figsize = (15,9))
plt.hist(stock_df['High_Moderna'], num_bins, facecolor = 'red')
plt.grid()
plt.title('Histogram: mu = ' + str(mu) + ', sigma = ' + str(sigma) )
```

# PLOT MULTIPLE PLOTS

```python
stock_df.reset_index().plot(x='Date', y=['High_Moderna', 'High_BioNTech', 'High_Johnson & Johnson'], figsize=(18,10), linewidth=4)
plt.ylabel('Price [$]')
plt.title('Vaccine Stock Prices for Moderna, BioNTech and J&J from 2019 - 2021')
plt.grid()
plt.legend(labels=['Moderna', 'BioNTech', 'J&J'], loc ='upper center')
```

# PLOT 3D PLOTS

```python
#import toolkits
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (15, 15))
ax = fig.add_subplot(111, projection = '3d')

x = stock_df['High_Moderna'].tolist()
y = stock_df['High_BioNTech'].tolist()
z = stock_df['High_Novavax'].tolist()


ax.scatter(x, y, z, c = 'r', s = 1000)
ax.set_xlabel('Moderna')
ax.set_ylabel('BioNTech')
ax.set_zlabel('Novavax')
ax.set_title('3D Plot - Vaccine Stock Prices for Moderna, BioNTech and Novavax from 2019 - 2021')
```
